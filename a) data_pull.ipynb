```python
import pandas as pd
from Stocks import Stock

import yaml
import sys
from pathlib import Path

def read_yaml(file_name): 
    with open(file_name) as file:
        yaml_data= yaml.safe_load(file)
    return yaml_data

config = read_yaml('config.yaml')
# config['canadian_etfs_parsed']
canadian_etfs_parsed = config['canadian_etfs_parsed']
```

```python
from datetime import datetime, timedelta

yesterday = (datetime.today() - timedelta(days=2)).strftime('%Y-%m-%d')
print(yesterday)
```

```python
dataframes_list = []

for x in canadian_etfs_parsed: 
    print(x)
    
    temp_data = Stock(x, '2018-03-01', yesterday).ticker_data
    temp_data['Stock_name'] = x
    dataframes_list.append(temp_data)
```

```python
yesterday= yesterday.replace("-", "_")
yesterday
```

```python
pd.concat(dataframes_list).to_parquet(f"ETFs_data_{yesterday}.parquet.gzip", compression='gzip')
```

```python
df = pd.read_parquet(f"ETFs_data_{yesterday}.parquet.gzip")
df.head()
```

```python
[x for x in list(set(df['Stock_name'])) if 'XEQ' in x]
```

```python
successfully_pulled_stocks = set(df['Stock_name'])
success_rate = len(successfully_pulled_stocks)/len(canadian_etfs_parsed)
success_rate
```

```python
failed_pulls = list(set(canadian_etfs_parsed) - set(successfully_pulled_stocks))
failed_pulls

failed_pulls_dataframes_list = []

for x in failed_pulls: 
    print(x)
    
    temp_data = Stock(x, '2018-03-01', '2023-03-01').ticker_data
    temp_data['Stock_name'] = x
    failed_pulls_dataframes_list.append(temp_data)
    
```
